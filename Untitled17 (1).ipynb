```python
import pandas as pd
import numpy as np
from sklearn.neighbors import LocalOutlierFactor

# Загрузка данных
data = pd.read_excel('train.xlsx')

# Предположим, что данные находятся в столбце 'Цена на арматуру'
prices = data['Цена на арматуру'].values.reshape(-1, 1)

# Инициализация модели LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)

# Предсказание выбросов
outliers = lof.fit_predict(prices)

# Фильтрация выбросов
filtered_data = data[outliers == 1]

# Сохранение очищенных данных в новый файл
filtered_data.to_excel('filtered_train.xlsx', index=False)

print("Выбросы удалены и данные сохранены в 'filtered_train.xlsx'")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as scs
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

data = pd.read_excel('filtered_train.xlsx')

data['dt'] = pd.to_datetime(data['dt'])

data.set_index('dt', inplace=True)

def invboxcox(y, lmbda):
    if lmbda == 0:
        return np.exp(y)
    else:
        return np.exp(np.log(lmbda * y + 1) / lmbda)

data['Цена на арматуру_box'], lmbda = scs.boxcox(data['Цена на арматуру'] + 1)

plt.figure(figsize=(12, 6))
plt.plot(data['Цена на арматуру_box'], label='Преобразованный ряд (Бокс-Кокс)')
plt.title('Ряд после преобразования Бокса-Кокса')
plt.xlabel('Дата')
plt.ylabel('Цена на арматуру (Бокс-Кокс)')
plt.legend()
plt.grid(True)
plt.show()

def tsplot(y, lags=None, figsize=(12, 7)):
    plt.figure(figsize=figsize)
    layout = (2, 2)
    ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
    acf_ax = plt.subplot2grid(layout, (1, 0))
    pacf_ax = plt.subplot2grid(layout, (1, 1))

    y.plot(ax=ts_ax)
    ts_ax.set_title('Временной ряд')
    plot_acf(y, lags=lags, ax=acf_ax)
    plot_pacf(y, lags=lags, ax=pacf_ax)
    plt.tight_layout()
    plt.show()

def check_stationarity(series):
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

check_stationarity(data['Цена на арматуру_box'])

data['Цена на арматуру_restored'] = invboxcox(data['Цена на арматуру_box'], lmbda)

plt.figure(figsize=(12, 6))
plt.plot(data['Цена на арматуру_restored'], label='Восстановленный ряд')
plt.title('Ряд после обратного преобразования Бокса-Кокса')
plt.xlabel('Дата')
plt.ylabel('Цена на арматуру')
plt.legend()
plt.grid(True)
plt.show()

data.to_excel('filtered_train_with_boxcox.xlsx')

print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

time_series = data['Цена на арматуру_box']

train_size = int(len(time_series) * 0.8)
train, test = time_series[:train_size], time_series[train_size:]

p, d, q = 1, 1, 1

model = ARIMA(train, order=(p, d, q))
fitted_model = model.fit()

forecast = fitted_model.forecast(steps=len(test))

mae = mean_absolute_error(test, forecast)
mse = mean_squared_error(test, forecast)
rmse = np.sqrt(mse)

plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data', color='orange')
plt.plot(test.index, forecast, label='Forecast', color='green')
plt.legend()
plt.title('ARIMA Model with Manual (p, d, q): Actual vs Forecasted Values')
plt.show()

print(f"Manual parameters (p, d, q): ({p}, {d}, {q})")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```
